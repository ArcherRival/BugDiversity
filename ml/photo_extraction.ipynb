```python
new_df = df.merge(oc_df, on="gbifID", how='left')
```

```python
new_df = new_df[new_df.vernacularName.notnull()]
```

```python
new_df = new_df.reset_index()
```

```python
image_df = new_df[new_df["vernacularName"].map(new_df["vernacularName"].value_counts()) >= 100].reset_index()
```

```python
image_df = image_df.filter(items=["gbifID","vernacularName","identifier"])
```

```python
def replaceSpaces(name):
    return name.replace(" ", "_")   
```

```python
image_df_subset = image_df[:1000]
image_df_subset["vernacularName"] = image_df_subset["vernacularName"].apply(replaceSpaces)
image_df_subset.to_csv("image_subset.csv")
```

```python
import requests
import os
def download(row):
    url = row["identifier"]
    id = row["gbifID"]
    name = row["vernacularName"]
    path = f'/home/chavez/Desktop/Software/Projects/BugDiversity/ml/images/{name}'
    img_data = requests.get(url).content
    if not  os.path.exists(path):
        os.makedirs(path)
    with open(f'/home/chavez/Desktop/Software/Projects/BugDiversity/ml/images/{name}/{id}.jpg', 'wb') as handler:
        handler.write(img_data)
```
