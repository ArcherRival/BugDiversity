```python
import pandas as pd
```

```python
df = pd.read_csv("./data/multimedia.txt", sep='\t')
df.describe()
```

```python
df.head()
```

```python
df = df.filter(items = ["gbifID","identifier"])
```

```python
oc_df = pd.read_csv("./data/occurrence.txt", sep='\t')
```

```python
oc_df = oc_df.filter(items=["gbifID","scientificName","vernacularName"])
```

```python
oc_df[oc_df.vernacularName.notnull()]
```

```python
new_df = df.merge(oc_df, on="gbifID", how='left')
```

```python
new_df = new_df[new_df.vernacularName.notnull()]
```

```python
new_df = new_df.reset_index()
```

```python
image_df = new_df[new_df["vernacularName"].map(new_df["vernacularName"].value_counts()) >= 100].reset_index()
```

```python
image_df = image_df.filter(items=["gbifID","vernacularName","identifier"])
```

```python
temp = image_df.groupby("vernacularName")
df_list  = [temp.get_group(x)[:20] for x in temp.groups]
```

```python
image_subset = pd.concat(df_list).reset_index()
```

```python
image_subset = image_subset.drop(columns=["index"])
image_subset
```
